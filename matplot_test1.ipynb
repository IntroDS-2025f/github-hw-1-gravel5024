````python
# ‰Ω†ÁöÑÁ®ãÂºè

import numpy as np
import matplotlib.pyplot as plt

# Âõ∫ÂÆö‰∫ÇÊï∏Á®ÆÂ≠ê‰ª•‰æøÁµêÊûúÂèØÈáçÁèæ
np.random.seed(768)
n=200

# Êï£‰ΩàÂúñË≥áÊñô
x = np.random.normal(loc=2.0, scale=1.0, size=n)
y = x * 0.5 + np.random.normal(scale=0.8, size=n)
size_var = np.abs(np.random.normal(loc=20, scale=10, size=n))  # Âü∫Á§éÈªûÈù¢Á©çÔºàÂèØÈÅ∏ÊìáÁ∏ÆÊîæÔºâ

# ÁÇ∫‰∫ÜË¶ñË¶∫ÊïàÊûúÔºåÂ∞ç size_var ÂÅöÈÅ©Â∫¶Á∏ÆÊîæÔºàmatplotlib ÁöÑ s ÂèÉÊï∏ÊòØÈù¢Á©çÔºâ
size_scale = size_var * 5

# Áï´Êï£‰ΩàÂúñ
plt.figure(2) # Create a new figure for the second plot
# c=y: Áî® y ÂÄºÊò†Â∞ÑÈ°èËâ≤
# s=size_scale: Áî®Á∏ÆÊîæÂæåÁöÑ size_var Ë®≠ÂÆöÈªûÁöÑÂ§ßÂ∞è
# cmap='viridis': ‰ΩøÁî® 'viridis' È°èËâ≤ map
# alpha=0.7: Ë®≠ÂÆöÈÄèÊòéÂ∫¶
scatter = plt.scatter(x, y, c=y, s=size_scale, cmap='viridis', alpha=0.7, edgecolors='none')

# Âä†‰∏ä colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('y-values')

# Ë®≠ÂÆö title
plt.title('Scatter with Colormap')

# È°ØÁ§∫ÂúñË°®
plt.show()
````
