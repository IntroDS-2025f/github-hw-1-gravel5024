## 📌 題目 1：直方圖 (20 分)

從正態分布產生 1000 筆資料，畫 **bins=30** 的直方圖。

其他要求: 

> title: Normal Distribution Histogram

````python
# 你的程式

import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title('Normal Distribution Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
````

````python
# 你的程式

import numpy as np
import matplotlib.pyplot as plt

# 固定亂數種子以便結果可重現
np.random.seed(768)
n=200

# 散佈圖資料
x = np.random.normal(loc=2.0, scale=1.0, size=n)
y = x * 0.5 + np.random.normal(scale=0.8, size=n)
size_var = np.abs(np.random.normal(loc=20, scale=10, size=n))  # 基礎點面積（可選擇縮放）

# 為了視覺效果，對 size_var 做適度縮放（matplotlib 的 s 參數是面積）
size_scale = size_var * 5

# 畫散佈圖
plt.figure(2) # Create a new figure for the second plot
# c=y: 用 y 值映射顏色
# s=size_scale: 用縮放後的 size_var 設定點的大小
# cmap='viridis': 使用 'viridis' 顏色 map
# alpha=0.7: 設定透明度
scatter = plt.scatter(x, y, c=y, s=size_scale, cmap='viridis', alpha=0.7, edgecolors='none')

# 加上 colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('y-values')

# 設定 title
plt.title('Scatter with Colormap')

# 顯示圖表
plt.show()
````
